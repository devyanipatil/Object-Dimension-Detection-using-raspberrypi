```python
import torch
from ultralytics.yolov5.models import Model
import cv2

# Load the model
model = Model("best.pt")

# Open the video stream
cap = cv2.VideoCapture(0)

while True:
    # Read a frame from the video stream
    ret, frame = cap.read()

    # Perform object detection on the frame
    outputs, _, _ = model(frame)

    # Extract the confidence scores from the output
    scores = outputs[0][:, 5]

    # Save the scores in a variable
    confidence_scores = scores.tolist()
    print(confidence_scores)

    # Draw the detections on the frame
    for x1, y1, x2, y2, conf, cls in outputs[0]:
        cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)
        cv2.putText(frame, f"{cls} {conf:.2f}", (x1, y1), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)

    # Show the frame
    cv2.imshow("Frame", frame)

    # Break the loop if the 'q' key is pressed
    if cv2.waitKey(1) == ord("q"):
        break

# Release the video stream and close the window
cap.release()
cv2.destroyAllWindows()
```

```python
import torch
import cv2

# Load the model with custom weights
model = torch.jit.load('C:/Users/prathamesh/Desktop/yolov5-master/runs/train/exp/weights/best.pt')
model.eval()

# Open the webcam
cap = cv2.VideoCapture(0)

while True:
    # Capture a frame
    ret, frame = cap.read()

    # Convert the frame to a PyTorch tensor
    x = torch.from_numpy(frame).float()
    x = x.permute(2, 0, 1)
    x = x.unsqueeze(0)
    x = x / 255.

    # Perform object detection
    with torch.no_grad():
        output = model(x)
    output = output[0]

    # Extract the confidence scores
    confidences = output['scores'].numpy()

    # Do something with the confidences
    print(confidences)
    
    # Display the frame
    cv2.imshow('frame', frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the webcam
cap.release()
cv2.destroyAllWindows()
```
